```python
import matplotlib.pyplot as plt
```

```python
x=[0,1,2]
y=[0,1,2]
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Straight line')
plt.show()
```

```python
plt.close()
```

```python
x1=[10,20,30]
y1=[10,30,40]
x2=[10,30,40]
y2=[25,10,30]
plt.plot(x1,y1,label='Line 1')
plt.plot(x2,y2,label='Line 2')
plt.legend
plt.show
```
